```python
import numpy as np

swiss_roll = np.loadtxt("../../data/2_class_swiss_roll.csv", delimiter=',')
labels = np.loadtxt("../../data/2_class_swiss_roll_labels.csv", delimiter=',')
```

```python
print swiss_roll.shape
```

```python
%matplotlib inline
import matplotlib.pylab as plt
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
inds1 = labels == 0
inds2 = labels == 1
ax.plot(swiss_roll[inds1, 0], swiss_roll[inds1, 1], swiss_roll[inds1, 2], 'rx')
ax.plot(swiss_roll[inds2, 0], swiss_roll[inds2, 1], swiss_roll[inds2, 2], 'bo')
```

```python
import sys
sys.path.append('..')
from swissy import laplacian_eigenmaps

unrolled = laplacian_eigenmaps.reduce(swiss_roll, d=2, k=10)
```

```python
plt.plot(unrolled[inds1, 0], unrolled[inds1, 1], 'rx')
plt.plot(unrolled[inds2, 0], unrolled[inds2, 1], 'bo')
```

```python
from sklearn.manifold import SpectralEmbedding
```

```python
se = SpectralEmbedding(n_components=2, n_neighbors=7)
unrolled1 = se.fit_transform(swiss_roll)
```

```python
plt.plot(unrolled1[inds1, 0], unrolled1[inds1, 1], 'rx')
plt.plot(unrolled1[inds2, 0], unrolled1[inds2, 1], 'bo')
```
